```python
import os 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_data = {}
windSpeed = {}
directory = os.getcwd()

# Read datasets
files = os.listdir(f"{directory}/data/raw")

for file in files:
    df = pd.read_csv(f"{directory}/data/raw/{file}")
    timeSerieWS = df['wind_speed_10m']

    all_data[file] = df
    windSpeed[file] = timeSerieWS

    plt.figure(figsize=(30, 3))
    plt.plot(timeSerieWS)
    plt.title(file)
    plt.xlabel('Time')
    plt.ylabel('Wind Speed')
    plt.xlim(0, len(timeSerieWS))
    plt.show()
```
